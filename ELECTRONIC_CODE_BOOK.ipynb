```python
div_16 = []
for i in range(0, len(pt_para), 16):
  div_16.append(pt_para[i:i+16])
div_16
```

```python
def to_ascii(div_16):
  ascii_pt = [] 
  for i in div_16: 
    ascii_pt_2 = []
    for j in i:
      ascii_pt_2.append(ord(j))
    ascii_pt.append(ascii_pt_2)
  return ascii_pt
```

```python
ascii_pt = to_ascii(div_16)
```

```python
def to_binary(ascii_pt):
  binary_pt = []
  for i in ascii_pt:
    binary_pt_2 = [] 
    for j in i: 
      binary_pt_2.append((bin(j)[2:].zfill(8))) 
    binary_pt.append(binary_pt_2)
  return binary_pt
```

```python
binary_pt = to_binary(ascii_pt)
print(binary_pt)
```

```python
if len(binary_pt[-1])!=16:
  for i in range(16-len(binary_pt[-1])):
    binary_pt[-1].insert(0, '00000000')
print(binary_pt)
```

```python
def XOR(binary_pt_2, mystr2): 
  final_p = [] 
  for i in binary_pt_2:
    p = []
    for j in range(len(i)): 
      p.append((int(i[j])^int(mystr2[j])))
    # print(p)
    final_p.append(p)
  return final_p
```

Left Shift

```python
def left_shift(p, t): 
  ls = []
  for i in p: 
    mystr = ""
    for k in i: 
      mystr += str(k) 
    ls.append(mystr[t:] + mystr[:t])
  return ls
```

```python
ls = left_shift(p, 1)
ls
```

```python
final_et = []
for i in ls:
  et = []
  while(i!=''): 
    et.append((i[:8])) 
    i = i[8:] 
  final_et.append(et)

encrypted = '' 
for i in final_et: 
  for j in i: 
    temp = int(j, 2) 
    encrypted += chr(temp)

print(encrypted) 
```
